11. Move the target variable to the front of df by running the cell with the following code:

This is done to distinguish the target from our features by storing it to the front of our DataFrame.
Here, we introduce a dummy variable y to hold a copy of the target column before removing it from the DataFrame. We then use the Pandas concatenation function to combine it with the remaining DataFrame along the 1st axis (as opposed to the 0th axis, which combines rows).

```python
# Move the target variable to the front of the table
y = df.MedHouseVal.copy()
del df['MedHouseVal']
df = pd.concat((y, df), axis=1)

# axis=0 means rows, concatenate vertically
# axis=1 means columns, concatenate horizontally
```

Note

1. run the cell containing the following sample DataFrames:

Here, we will build two simple DataFrames from scratch. 

As can be seen, they contain a product column with some shared entries.

```python
# Build a simple set of tables

df_1 = pd.DataFrame({'product': ['red shirt', 'red shirt', 'red shirt', 'white dress'],
                     'price': [49.33, 49.33, 32.49, 199.99]})
df_2 = pd.DataFrame({'product': ['red shirt', 'blue pants', 'white tuxedo', 'white dress'],
                     'in_stock': [True, True, False, False]})
```

2. Run the next cell to perform the inner merge:

```python
# Inner merge

df = pd.merge(left=df_1, right=df_2, on='product', how='inner')
df
```

Note 
---
only the shared items, red shirt and white dress, are included. To include all entries from both tables, we can do an outer merge instead. Let's do this now

3. Run the next cell to perform an outer merge:

```python
# Outer merge

df = pd.merge(left=df_1, right=df_2, on='product', how='outer')
df
```

3.  Handling with MissingValues:

 A very common preprocessing step that any scientist perform before fitting the model is the so called Imputation of the missing values.

 Typical one has two strategies: 
 1. Remove the example that shows one (or more) missing values; 
 2. Impute the missing value with a reasonable summary statistics

 1. Remove the example that shows one (or more) missing values; 


```python
# Drop the incomplete rows

df.dropna()
```

```python
# Drop the incomplete columns

df.dropna(axis=1)
```

 2. Impute the missing value with a reasonable summary statistics:

 a. use DataFrame method for filling missing values;

 b. use SimpleImputer class.

 ##### a. use DataFrame method for filling missing values;

```python
## read the df again to see the effect:
df = pd.merge(left=df_1, right=df_2, on='product', how='outer')
df
```

```python
# Fill with the average

df.fillna(value=df.price.mean())
```

```python
# Fill with the previous value in that column

df.fillna(method='pad')
```

Notice how the white dress price was used to pad the missing values below it.

 ##### b. use SimpleImputer class:

1- First initialize an instance of the SimpleImputer class by indicating the strategy (mean) as well as specifying the missing values that you want to locate (np.nan):

2- Once the instance is created, you use the fit() function to fit the imputer on the column(s) that you want to work on.

3- Use the transform() function to fill the missing values based on the strategy you specified in the initializer of the SimpleImputer class.


```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean', missing_values=np.nan)

## read the df again to see the effect:
df = pd.merge(left=df_1, right=df_2, on='product', how='outer')
df

```

```python

imputer = imputer.fit(df[['price']])
df['price'] = imputer.transform(df[['price']])
df
```

#### Building training data sets

1. Run the first cell in the Building training data sets section to add another column of data representing the average product ratings before encoding the labels:

```python
df = df.fillna(value=df.price.mean())
ratings = ['low', 'medium', 'high']
np.random.seed(2)
df['rating'] = np.random.choice(ratings, len(df))
df
```

2. Dealing with Categorical Variables:

Considering we want to use this table to train a predictive model, we should first think about changing all the variables to numeric types.

a- Convert the handle in_stock., which is a Boolean list, to numeric values; 

for example, 0 and 1. 

This should be done before using it to train a predictive model.
This can be done in many ways, for example, by running the cell containing the following code:

```python
# Convert in_stock to binary

df.in_stock = df.in_stock.map({False: 0, True: 1})
df
```

b- Run the cell containing the following code to map class labels to integers at a higher level. 

We use sci-kit learn's LabelEncoder for this purpose:


Here, we instantiate a label encoder and then "train" it and "transform" our data using the fit_transform method. We apply the result to a copy of our DataFrame, _df.

```python
# Encode ordinal labels

from sklearn.preprocessing import LabelEncoder
rating_encoder = LabelEncoder()
df.rating = rating_encoder.fit_transform(df.rating)
df
```

You may notice a problem here. We are working with a so-called "ordinal" feature, where there's an inherent order to the labels. In this case, we should expect that a rating of "low" would be encoded with a 0 and a rating of "high" would be encoded with a 2. However, this is not the result we see. In order to achieve proper ordinal label encoding, we should again use map, and build the dictionary ourselves.

c- Re-convert the features using the class we reference with the variable rating_ encoder, by running rating_encoder.inverse_ transform(df.rating):

```python
# Convert back if needed

rating_encoder.inverse_transform(df.rating)
```

d- Encode the ordinal labels properly by running the cell containing the following code:

Convert the class labels from strings to numerical values by running the cell containing the following code:

```python
# One-hot-encode the product feature

df = pd.get_dummies(df)
df
```

Here, we see the result of one-hot encoding: the product column has been split into 4, one for each unique value. Within each column, we find either a 1 or 0 representing whether that row contains the particular value or product.
