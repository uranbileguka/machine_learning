### Step to do:

        1. Import the necessary packages from scikit-learn (KMeans, AgglomerativeClustering, and silhouette_score). (1 points)

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import pandas as pd
import matplotlib.pyplot as plt
```

        2. Read the Mall Customers dataset into the pandas DataFrame and display the last 10 samples. (1 points)

```python
mc_df = pd.read_csv("Mall_Customers.csv")
mc_df.tail(10)
```

       3. Assign the two features Annual Income (k$)" and "Spending Score (1-100)" to X. (2 points)

```python
X = mc_df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
X.head()
```

        4. Visualize the the two featuerse ("Annual Income (k$)" And "Spending Score (1-100)) to understand the data structure. (2 points)

```python
plt.figure(figsize=(8,6))
plt.scatter(mc_df['Annual Income (k$)'], mc_df['Spending Score (1-100)'], color='teal')
plt.title('Customer Distribution by Income and Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```

        6. implementation of k-means on the X features using K= 5. ( 2 points)

```python
km = KMeans(n_clusters=5, random_state=42)
km_clusters = km.fit_predict(X)
```

        7. Plot the predicted clusters from k-means. ( 2 points)

```python
plt.figure(figsize=(10,6))
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'], c=km_clusters)
plt.title('K-Means Clustering of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

        8. compute and print the silhouette score of K-mean clustering method (2 point)

```python
print("K-Means Clustering, silhouette score: ", silhouette_score(X, km_clusters))
```

        9. Implementation of Agglomerative Clustering on the X.

You will implement the Agglomerative Clustering three times using three types of linkages (single, complete and average)

#### a - single linkages:
* Implementation of Agglomerative Clustering with single linkages,  (1 point)
* Plot the predicted clusters. (1 point)
* compute the silhouette score  (1 point)

```python
# Implementation of Agglomerative Clustering with single linkages
ac_s = AgglomerativeClustering(5, linkage='single')
ac_s_clusters = ac_s.fit_predict(X)
```

```python
# Plot the predicted clusters
plt.figure(figsize=(10,6))
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'], c=ac_s_clusters)
plt.title('Agglomerative Clustering of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
# compute and print the silhouette score
print("Silhouette score: ", silhouette_score(X, ac_s_clusters))
```

```python
# Implementation of Agglomerative Clustering with complete linkages
ac_c = AgglomerativeClustering(5, linkage='complete')
ac_c_clusters = ac_c.fit_predict(X)
```

```python
# compute and print the silhouette score
print("Silhouette score: ", silhouette_score(X, ac_c_clusters))
```
